# Machine Learning Landscape
How do we know which models to use?   

3 different ways to categorize ML Algorithms
1. Supervision (Labeled data, Unlabeled like addresses, or Semilabeled data like when some data has labels and others don't)
2. Online vs Batch Learning
    - Online: learns as it gathers data. For example, recommendation systems like opening YouTube for the first time.
    - Batch Learning: Give it data, learn from it, then use that model for the future data
3. Instance-Based vs Model Based learning
    - How it uses the new data. Does it need to create a new model every time the data changes?
**not exclusive models, but some of them are these models**


## Supervision  
- Supervised - labeled data
    - discrete data (categorical models) --> "classification models"
    - continuouse data (quantitative models) ==> "Regression models"
- Unsupervised - unlabeled data
    - discrete data (categorical models) --> "clustering models"
    - continuouse data (quantitative models) ==> "embedding/ dimension reduction/ visualization/ anomaly detection/ association rule learning/ ... models"
- SemiSupervised


## Online vs Batch Learning  

### Batch
Batch Learning: All the data is provided at the beginning and model is trained only on this batch of data  

Pros:   
- Consistency - changes not needed
- type of data
- base model to train other models on
- automated (low maintenance)


Cons:  
- old and invalid
- inflexible
- data is needed upfront
- storage intensive (if automated, it needs to carry new data with it)
- Time-intensive  

### Online learning (Incremental learning)   
Train the system incrementally by feeding it new data instances.  
*Learning Rate*: The speed at which the system adapts.  High rates adapt quickly, but also forget old data quickly. Slow rates will cause the model to learn slower, but will have less noise in the data.

Pros:
- More up-to-date
- Model improves over time
- Flexible
- Less data needed upfront

Cons:  
- Innacurate at first
- Model might be too large
- Bad data will ruin it 


## Instance Based Learning vs. Model Based Learning  
### Instance  
System learns the examples, then generalizes  . 
Ex. show 100 pictures of a cat and of a dog, then learn from those features.  

Pros:  
- Simple to classify something if it is a category
Cons:  
- Can easily overfit

### Model-based Learning  
Build a model from examples, then use the model to make predictions.  
Ex. This image has pointed ears, and the majority of dogs have pointed ears, so it is a dog  

Pros:  
- Make multiple predictions (what's the most likely category)
Cons:
- Not binary output.
