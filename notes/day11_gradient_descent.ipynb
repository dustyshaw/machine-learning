```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_set = np.linspace(-4.5, )
def Loss(w):
    return w**3 + 4*w**2

x_set = np.linspace(-4.5, 1.5, 100)
L_set = Loss(x_set)

plt.plot(x_set, L_set)
```

1. Make a model
$$f(x) = b_0 + b_1x$$
$$f(x) = b_0 - b_1x$$


2. Establish loss Function
3. Find the slope of the Loss Function $\frac{\partial L}{\partial b_1}$
 - Partial derivatives are only taking the derivative with respect to that variable, other variables are counted as constants
4. Take our weight and subtract the slope.

$$ b_1 = b_1 = \eta \frac{\partial L}{\partial b_1} $$

blah = learning rate. Should be something small $\eta$



```python
b1 = 1 # starting point
Lb1 = Loss(b1) # Loss function of b1

# Choose the learning rate
eta = 0.05

# Gradient Descent
plt.plot(x_set, L_set, alpha=0.1)
plt.plot(b1,Lb1,'ro')

numSteps = 19

for i in range(numSteps):
    # evaluate the loss function, then at a nearby, then divide by the distance h (defenition of a derivative)
    
    h = 0.01
    dLdb = (Loss(b1+h) - Loss(b1)) / h
    b1 = b1 - eta*dLdb
    plt.plot(b1, Loss(b1), 'ro', alpha=(i+1)/numSteps)
```
