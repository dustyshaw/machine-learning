# Bigram

Consider this fairy tail:
> Once upon a time, there was a duck...

1. Create an entire list of unique words
2. for each word, make a list of words that follow
3. Now, for each word, create a probability distribution of the words that follow  


2nd word: Probability  
'upon' : 8%  
'and' : 34%  
'there' : 22%  
'.': 24%  

total: 100%


4. Model: Randomly select a starting word.
5. Randomly select the word to follow the first word based on porbability. 
    * Choose a random number 0-100, if it is 0-8, use "upon" for example.
6. Move to the newly selected word and repeat step 5.
7. Repeat until the new word is selected as a '.'
    * can continue with another sentence if wanted

````python
# biGram (2-Gram)

list2 = {
    "once": ["upon", "and", ".", "but"],
    "upon": ["a", "another"],
    "a": ["time", "duck", "dog", "reflection", "swan"],
    "time": ["there", ".", "to", "there"],
    "there": ["was", ".", "he", "is", "were", "."]
}
````

````python
import pandas as pd
import re
from numpy import random

def fix_unicode(text: str) -> str:
    return text.replace(u"\u2019", "'")

import re
from numpy import random
import requests

url = 'https://raw.githubusercontent.com/drolsonmi/math3480/refs/heads/main/datasets/WhatIsDataScience.txt'
response = requests.get(url)
content = response.text.split('\r\n')


````

````python
content
````

````python
regex = r"[\w']+|[\.]"

document = []

for paragraph in content:
    if not paragraph == "":
        words = re.findall(regex, fix_unicode(paragraph))
        document.extend(words)
        
document
````

````python
from collections import defaultdict

transitions =  defaultdict(list) # list of unique characters iwth list that follows each word

# look through the document at every enty within document and the word right after it

for previous, current in zip(document, document[1:]): 
    transitions[previous].append(current)
````

````python
def generate_using_bigrams() -> str:
    current = '.'
    result = []
    while True:
        next_word_candidates = transitions[current]
        current = random.choice(next_word_candidates)
        result.append(current)
        if current == "." : return " ".join(result)
````

````python
generate_using_bigrams()
````

## Trigram  
1. Randomly choose a first word
2. Choose a 2nd word as before
3. Choose a 3rd word to follow the 1st and 2nd words

```python
 list={
    ["once", "upon"] : "a",
    ["upon", "a"] : "time"
 }
```

4. Select [2nd and 3rd] words and choose the 4th word
5. select [word-i and word-(i+1)] and choose word-(i+2)
6. Continue this process until the sentence or paragraph is completed


````python
trigram_transitions = defaultdict(list)

starts = []

for previous, current, next in zip(document, document[1:], document[2:]): 
    if previous == '.':
        starts.append((current,next))
    trigram_transitions[(previous, current)].append(next)
````

````python
import numpy as np

def generate_using_trigrams() -> str:
    current = random.choice(starts)
    prev = "."
    result = [current]
    while True:
        next_word_candidates = trigram_transitions[(prev,current)]
        next_word = random.choice(next_word_candidates)
        prev, current = current, next_word
        result.append(current)
        if current == "." : return " ".join(result)
````

````python
generate_using_trigrams()
````

# Grammars

We build a sentence by decomposing the sentence structure. 

Ex. sentence `_S`. A sentence is made of a Noun phrase and a Verb phrase.  
A noun phrase can be broken down futher, for example, it can be...
    * a noun
    * an adjective + a noun 
    * an adjective + noun phrase + a preposition + adjective + noun
    * etc...  


A verb phrae can be broken...
    * a verb
    * a verb + noun phrase 


* `_NP` = Noun Phrase --> [`_N`] or [`_A`, `_N`]
* `_VP` = Verb Phrase --> [`_V`] or [`_V`, `_NP`]
* `_N`
* `_A`
* `_P`
* `_V`  

Start with a Sentence:
* [`_S`]  
This sentence "decomposes" into a noun phrase and a verb phrase:  
* [`_NP`, `_VP`]

A Noun Phrase decomposes into (make a random choice)
* [`_A`,`_N`,`_VP`]

then...
* [`_A`,`_N`,`_V`, `_NP`]  

then... `_NP` ->  `_A`,`_N`
* [`_A`,`_N`,`_V`, `_A`,`_N`]   

Fill each noun, adjective, preposition, and verb with words from a pre-determined list.




````python
from typing import List, Dict

# Type alias is going to refer to grammars later

Grammar = Dict[str, List[str]] # ex `_N`, ("Dusty", "duck", "science"...)

grammar = {
    "_S": ["_NP _VP"],
    "_NP": ["_N", "_A _N", "_A _NP _P _A _N"],
    "_VP": ["_V", "_V _np"],
    "_N": ["data science", "Python", "regression", "Dusty", "Enoch"],
    "_A": ["big", "linear", "logistic", "crazy", "logical"],
    "_P": ["about", "near", "to", "over", "around"],
    "_V": ["learns", "trians", "is", "tests", "has", "creates", "builds", "loves"]
}

def is_terminal(token: str) -> bool:
    return token[0] != "_"

def expand(grammar: Grammar, tokens: List[str]) -> List[str]:
    for i, token in enumerate(tokens):
        if is_terminal(token): continue
        
        replacement = random.choice(grammar[token])
        
        if is_terminal(replacement):
            tokens[i] = replacement
        else:
            tokens = tokens[:i] + replacement.split() + tokens[(i+1):]
            
        return expand(grammar, tokens)
    return tokens


def generate_sentence(grammar: Grammar) -> List[str]:
    return expand(grammar, ["_S"])
````
