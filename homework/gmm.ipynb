```python
import pandas as pd
data = pd.read_csv("./Data/clustering_data.csv")
```

```python
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import seaborn as sns

# Optional: name the columns for easier reference
data.columns = ['x', 'y']

# Fit Gaussian Mixture Model
gmm = GaussianMixture(n_components=6, random_state=0)  # Change n_components as needed
gmm.fit(data)

# Predict cluster labels
labels = gmm.predict(data)

# Add labels to the dataframe
data['Cluster'] = labels

# Plot the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Cluster', palette='viridis')
plt.title("GMM Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# KDE jointplot with cluster hues
sns.jointplot(data=data, x="x", y="y", hue="Cluster", kind="kde", fill=True, palette="viridis")

# plt.suptitle("GMM Clustering - KDE Joint Plot", y=1.02)
plt.show()
```
